Crear una función que muestre los siguientes gráficos para las variables numéricas según una variable cualitativa:

Histograma con Curva


```python
df_num = df.loc[:, ['Peso_En_Libras', 'Desplazamiento', 'Aceleracion']] 
df_num.head(6)
```

```python
def grafico_numericos(columnas, data, x): # parámetros
  
  for i, colum in enumerate(columnas): 
    fig, ax = plt.subplots(figsize=(6, 3.84))
    sns.boxplot(   # boxplot  violinplot
        x = x,  # eje X
        y = colum,  # eje Y
        data  = data,
        )
    ax.set_title(colum, fontsize = 10, fontweight = "bold")
    ax.tick_params(labelsize = 8)
    ax.set_xlabel("Nivel")
```

```python
#Grafico de boxplot por tipo de cilindro
x = df.Cilindros  # variable cualitativa
grafico_numericos(columnas, df_num, x)
```
